```python
import numpy as np
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests

# Initialize lists
prices = []
brands = []
update = []
dis = []  

# Define the base URL for scraping
urls = "https://www.flipkart.com/search?q=chains&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

# Loop through 10 pages
for i in range(1, 11):
    pages = urls + "&page=" + str(i)
    page = requests.get(pages)
    soup = BeautifulSoup(page.content, "html.parser")

    # Prices
    p = soup.find_all("div", class_="Nx9bqj")
    for item in p:
        if len(prices) >= 400:
            break
        text = item.get_text(strip=True)
        clean_price = text.replace("₹", "").replace(",", "")
        prices.append(clean_price)

    # Brands
    b = soup.find_all("div", class_="syl9yP")
    for item in b:
        text = item.text.strip()
        brands.append(text if text else np.nan)

    # Titles
    m1 = soup.find_all("a", class_="WKTcLC")
    for i in m1:
        title = i.text.strip().replace("\n", " ")
        update.append(title)

    
    m2 = soup.find_all("div", class_="UkUFwK")
    for i in m2:
        if len(dis) >= 400:
            break
        a = i.text.strip()
        a = re.sub(r"\s*\%?\s*off\s*$", "", a, flags=re.IGNORECASE)  # Remove 'xx% off'
        dis.append(a)

# Example patterns for different attributes
patterns = {
    "Plating": r"\b(?:Gold(?:en)?(?:[- ]Plated)?|Silver(?:[- ]Plated)?|Rose Gold|One Gram Gold Plated|1 Gram Gold Plated)\b",
    "Material": r"\b(?:Stainless Steel|Sterling Silver|Brass|Alloy|Copper|Wood|Titanium)\b",
    "Gender": r"\b(?:Men(?:'s)?|Women(?:'s)?|Boys|Girls|Unisex)\b",
    "Type": r"\b(?:Necklace|Chain|Pendant|Mala|Locket)\b",
    "Length": r"\b(?:\d{1,2}[\"']?\s?(?:in(?:ch)?|inches)?)\b",
    "Design": r"\b(?:Classic|Stylish|Designer|Valentine|Roman Numeral|Trendy|Fashionable)\b"
}
------------------------------------------------------------------------------------------------------------------------------------------------

def extract_info(title):
    extracted = {}
    for field, pattern in patterns.items():
        matches = re.findall(pattern, title, flags=re.IGNORECASE)
        extracted[field] = ", ".join(set(matches)) if matches else ""
    
    return extracted

# Extract structured data for each product title
structured_data = [extract_info(title) for title in update]

# Create DataFrame from the structured data
df1 = pd.DataFrame(structured_data)

# Show the DataFrame to check
print(df1)
```
